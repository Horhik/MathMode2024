### `1.` График $x^2⋅sin(1/x^2)+x$ на интервале $[−1,1]$

```python
def f(x):
    return x**2*np.sin(1/x**2) + x
    
start, stop, step = -1, 1, 250
x = np.linspace(start, stop, step)
y = f(x)

fig, ax  = plt.subplots()
ax.grid(True)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_title("$f(x) = x^2\sin{\\frac{1}{x^2}}+x$")

ax.plot(x,y)
```

### `2.` Semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы  функций $1/(1+x^2)$ и $1/x^2$ на интервале $[5,25]$. (Относительная разница чисел $a$ и $b$ определяется как $|1−a/b|$)


```python

def f(x):
    return 1/(1 + x**2)
def g(x):
    return 1/x**2

start, stop, step = 5, 25, 500
x = np.linspace(start,stop, step)
fy = f(x)
fg = g(x)
y = fy/fg
y = 1 - y

fig, ax  = plt.subplots()
ax.grid(True)

ax.set_xlabel("X logarithmic axis")
ax.set_xscale("log")

ax.set_ylabel("Y linear axis")
ax.set_yscale("linear")

ax.set_title("$f(x) = 1 - \\frac{\\frac{1}{1 + x^2}}{\\frac{1}{x^2}} $", fontsize=20)

ax.plot(x,y)
```

### `4.` Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.Ы

```python
def f(x,y):
    return np.sin(x)*np.sin(y)
start, stop, step = -np.pi, np.pi, 500
x,y = np.meshgrid(np.linspace(start,stop, step), np.linspace(start,stop,step))
z = f(x,y)

fig, ax  = plt.subplots()
ax.grid(None)
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_title("$\\sin(x)\\sin(y)$")

ax.pcolormesh(x,y,z, cmap="inferno")
```

(подсказка: используйте готовые функции из библиотеки `np.linalg`).  

```python
from numpy import linalg, trace
A = np.random.rand(5,5)
B = A.T @ A
C = A @ A.T
print("A: \n", A)
print("B: \n", B)
print("C: \n", C)
print(f'\n det(A)={linalg.det(A)}\n det(B)={linalg.det(B)} \n det(C)={linalg.det(C)}\n')
print(f' trace(A)={trace(A)}\n trace(B)={trace(B)} \n trace(C)={trace(C)}\n')
print(f' matrix_rank(A)={linalg.matrix_rank(A)}\n matrix_rank(B)={linalg.matrix_rank(B)} \n matrix_rank(C)={linalg.matrix_rank(C)}\n')
print(f' eigvals(A)={linalg.eigvals(A)}\n eigvals(B)={linalg.eigvals(B)} \n eigvals(C)={linalg.eigvals(C)}\n')
```

```python
def compute_rect(a, b, f, d=0.1):
    steps = f(np.arange(a,b,d)) * d
    return np.sum(steps)
    
def compute_trap(a,b, f, d=0.1):
    steps = f(np.arange(a,b,d))
    steps = steps[0] + steps[-1] + 2*np.sum(steps[1:-1])
    steps = d/2*steps
    return steps
           
def compute_simpson(a,b,f, d=0.1):
    x_values = np.arange(a, b + d, d)
    y_values = f(x_values)
    result = d / 3.0 * (y_values[0] + 4 * np.sum(y_values[1:-1:2]) + 2 * np.sum(y_values[2:-2:2]) + y_values[-1])
    return result

def f(x):
    return (x**2 + 3)/(x-2)

print("For function #0")
delta = 1e-7
print("Rect: ", "{:.5f}".format(compute_rect(3,4, f, d = delta)))
print("Trap: ", "{:.5f}".format(compute_trap(3,4, f, d = delta)))
print("Simpson: ", "{:.5f}".format(compute_simpson(3,4, f, d = delta)))

def f(x):
    return np.cos(x)**3


print("For function #7")
delta = 0.000001
print("Rect: ", "{:.5f}".format(compute_rect(0, np.pi/2, f, d = delta)))
print("Trap: ", "{:.5f}".format(compute_trap(0, np.pi/2, f, d = delta)))
print("Simpson: ", "{:.5f}".format(compute_simpson(0, np.pi/2, f, d = delta)))
```
